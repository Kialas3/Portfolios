```python
from PIL import ImageGrab
import numpy as np
import cv2
import time

image = ImageGrab.grab()
width, height = image.size

fourcc = cv2.VideoWriter_fourcc(*'XVID')
video = cv2.VideoWriter('test.avi', fourcc, 25, (width, height))

# def drawLine(img,rho,theta):
#     a = np.cos(theta)
#     b = np.sin(theta)
#     x0, y0 = a * rho, b * rho
#     pt1 = (int(x0 + 1000 * (-b)), int(y0 + 1000 * a))
#     pt2 = (int(x0 - 1000 * (-b)), int(y0 - 1000 * a))
#     cv2.line(img, pt1, pt2, (255, 0, 0), 1, cv2.LINE_AA)

while True:
    start=time.time()
    img_rgb = ImageGrab.grab(bbox=(480, 415, 1300, 565))
    now_np_img=np.array(img_rgb)
    img_g=cv2.cvtColor(now_np_img, cv2.COLOR_RGB2GRAY)
    kernel_size = 5
    blur_gray = cv2.GaussianBlur(img_g,(kernel_size,kernel_size),0)
    img_canny=cv2.Canny(blur_gray,100,150)
    
    left_rho=0
    left_angle_sum=0
    left_line_num=0
    right_rho=0
    right_angle_sum=0
    right_line_num=0

    lines = cv2.HoughLines(img_canny, 1, np.pi / 180.0, 50)
    if lines is not None:
        a, b, c = lines.shape
        for i in range(a):
#             rho = lines[i][0][0]
            theta = lines[i][0][1]
            angle=theta*180/np.pi
            if(angle<80 and angle>10):
#                 left_rho+=rho
                left_angle_sum+=angle
                left_line_num+=1
            if(angle>100 and angle<170):
#                 right_rho+=rho
                right_angle_sum+=angle
                right_line_num+=1
#         if(left_line_num!=0):
#             drawLine(now_np_img,left_rho/left_line_num,left_angle_sum/left_line_num*np.pi/180)
#             print("left : "+str(left_angle_sum/left_line_num))
#         if(right_line_num!=0):
#             drawLine(now_np_img,right_rho/right_line_num,right_angle_sum/right_line_num*np.pi/180)
#             print("right : "+str(right_angle_sum/right_line_num))
    
    end=time.time()
    dt=end-start
    information=str(left_angle_sum)+" "+str(left_line_num)+" "+ str(right_angle_sum)+" "+str(right_line_num)+" "+str(dt)
    print(information)
    s1.send(bytes(information,"utf-8"))
    #cv2.imshow('imm', now_np_img)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
